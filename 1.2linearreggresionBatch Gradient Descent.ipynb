```python
class GDRegressor:
  def __init__(self,learning_rate,epochs):
    self.m=0
    self.b=0
    self.lr=learning_rate
    self.epochs=epochs

  def fit(self,X,y):
    for i in range(self.epochs):
      slopeb=-2*np.sum(y- self.m*X.ravel() - (self.b))
      slopem=-2*np.sum(X.ravel()*(y- self.m*X.ravel() - (self.b)))


      self.b=self.b- (self.lr*slopeb)
      self.m=self.m- (self.lr*slopem)

    print(self.b)
    print(self.m)
```
